```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
tickerData = yf.Tickers("AAPL MSFT TSM NVDA AVGO")
historyDf = tickerData.history(period="10y")
plt.figure(figsize=(16, 9),dpi=400)
plt.plot(historyDf["Close"])
plt.legend(historyDf.columns)
```

```python
sns.heatmap(historyDf["Close"].corr(method ='pearson'), annot=True, cmap="RdBu_r", linewidths=.5, linecolor="white")
```

```python
sns.heatmap(historyDf["Close"].corr(method ='kendall'), annot=True, cmap="RdBu_r", linewidths=.5, linecolor="white")
```

```python
sns.heatmap(historyDf["Close"].corr(method ='spearman'), annot=True, cmap="RdBu_r", linewidths=.5, linecolor="white")
```
